```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)


# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

#school_data_complete.hd viewing dataframe
school_data_complete.head()
```

```python
num_schools = school_data_complete["school_name"].unique()
number_schools = len(num_schools)
number_schools
#number of schools is 15
```

```python
number_students = school_data_complete["student_name"].count()
number_students
#number of students 39170
```

```python
total_budget = school_data_complete["budget"].unique().sum()
total_budget
```

```python
average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count/(number_students)*100
passing_math_percentage
```

```python
passing_reading_count =school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count/(number_students)*100
passing_reading_percentage
```

```python
passing_math_reading_count = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_both_percent = passing_math_reading_count /float(number_students)*100
passing_both_percent
```

```python
#create a high-level snapshot of the district's key metrics in DataFram
labels = [{"Total Schools": number_schools, "Total Students": number_students, "Total Budget": total_budget, "Avg Math Score": average_math_score,
           "Avg Reading Score": average_reading_score, "% Passing Math": passing_math_percentage, "% Passing Reading": passing_reading_percentage,
           "% Overall Passing":passing_both_percent}]

district_summary = pd.DataFrame(labels)


#format the output
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)



district_summary
```

```python
school_passing_math = school_data_complete[(school_data_complete["math_score"]) >= 70]
school_passing_math

school_passing_reading = school_data_complete[(school_data_complete["reading_score"]) >= 70]
school_passing_reading
```

```python
#identify the students that have passed both math and reading 
passing_math_and_reading = school_data_complete[(school_data_complete["reading_score"] >= 70)&(school_data_complete["math_score"] >= 70)]
passing_math_and_reading
```

```python
#calculate the passing rates 
per_school_passing_math = school_passing_math.groupby(["school_name"]).count()["student_name"] / per_school_student_count * 100
per_school_passing_reading = school_passing_reading.groupby(["school_name"]).count()["student_name"] / per_school_student_count * 100
overall_passing_rate = passing_math_and_reading.groupby(["school_name"]).count()["student_name"] / per_school_student_count * 100
overall_passing_rate
```
